```python
import pandas as pd
import numpy as np
import matplotlib as plt

path = 'Euro_2012_stats_TEAM.csv'
# Read the CSV file into a pandas DataFrame
euro12 = pd.read_csv(path, delimiter=',')

discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']]
print(discipline)

# Sort the DataFrame by 'Red Cards' and then 'Yellow Cards'
discipline_sorted = discipline.sort_values(by=['Red Cards', 'Yellow Cards'], ascending=False)

# Print the sorted DataFrame
print(discipline_sorted)

#Print the mean of yellow cards
print('the mean “Yellow Cards” given by Team is ',discipline_sorted['Yellow Cards'].mean())

#Filter teams scored with more 6 goals, using euro12 dataFrame
more_than_six_goals = euro12.loc[euro12['Goals'] > 6]
print('Only ',more_than_six_goals['Team'],'\n have scored more than 6 times')

# Filter teams that start with the letter 'G'
G_teams = euro12[euro12['Team'].str.startswith('G')]
print(G_teams['Team'])

# Display the first 7 columns
first_7_columns = euro12.iloc[:, :7]
print(first_7_columns)

# Display all columns except the last 3 columns
Num_col = len(euro12.columns)
all_except_last3 = euro12.iloc[:, :-3]
print(all_except_last3)
```
