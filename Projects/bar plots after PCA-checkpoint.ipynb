```python
import matplotlib.pyplot as plt
import numpy as np

# Results
models = ['Linear Regression', 'SVR', 'Decision Tree Regressor', 'Random Forest Regressor', 'Ridge Regression', 'Lasso Regression']
elapsed_times = [0.08876228332519531, 142.0030279159546, 0.9355249404907227, 8.294806718826294, 0.0807187557220459, 0.0797891616821289]
r2_scores = [0.8333942436117515, 0.9887321027008208, -0.21800622684243476, 0.9925509525047028, None, None]  # None for Ridge and Lasso

# Replace None values with a default value (e.g., 0)
default_r2_value = 0
default_elapsed_times_val = 0
r2_scores = [r2 if r2 is not None else default_r2_value for r2 in r2_scores]
elapsed_times = [e_t if e_t is not None else default_elapsed_times_val for e_t in elapsed_times]

# Filter models for SVR and Random Forest
selected_models = ['SVR', 'Random Forest Regressor']
selected_r2_scores = [r2 for model, r2 in zip(models, r2_scores) if model in selected_models]
selected_elapsed_times = [et for model, et in zip(models, elapsed_times) if model in selected_models]

# Plot R-squared scores
plt.figure(figsize=(8, 4))
plt.bar(selected_models, selected_r2_scores, color='red')
plt.title('R-squared Scores for Selected Regression Models')
plt.xlabel('Regression Models')
plt.ylabel('R-squared Score')
plt.xticks(rotation=45, ha='right')

# Zoom in on the y-axis values around 1
plt.ylim([0.98, 1.0])

plt.show()

# Plot elapsed times
plt.figure(figsize=(8, 4))
plt.bar(selected_models, selected_elapsed_times, color='blue')
plt.title('Elapsed Times for Selected Regression Models')
plt.xlabel('Regression Models')
plt.ylabel('Elapsed Time (seconds)')
plt.xticks(rotation=45, ha='right')
plt.show()
```
